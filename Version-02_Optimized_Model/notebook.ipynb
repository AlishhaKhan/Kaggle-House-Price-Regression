**📌 PART 06 — Data Cleaning & Missing Value Handling**

**📂 Step 1 — Make Copies of Dataset**

*Always keep the original dataset safe.*

```python
# Create copies of datasets

train_df = train.copy()
test_df = test.copy()
```

**📂 Step 2 — Drop Features with Extremely High Missing Values**

*Let's first inspect them.*

```python
# Features having more than 80% missing values

high_missing = missing_table[
    missing_table["Missing Percentage"] > 80
]

high_missing
```

*Now drop only those that are mostly missing and add very little information.*

```python
columns_to_drop = [
    "PoolQC",
    "MiscFeature",
    "Alley",
    "Fence"
]

train_df.drop(columns=columns_to_drop, inplace=True)
test_df.drop(columns=columns_to_drop, inplace=True)
```

**📂 Step 3 — Handle Numerical Missing Values**

*Instead of filling everything together, let's select numerical columns.*

```python
# Numerical columns except target variable
numeric_cols = train_df.select_dtypes(
    include=["int64", "float64"]
).drop(columns=["SalePrice"]).columns

for col in numeric_cols:

    train_df[col] = train_df[col].fillna(
        train_df[col].median()
    )

    test_df[col] = test_df[col].fillna(
        train_df[col].median()
    )
```

&&📂 Step 4 — Handle Categorical Missing Values**

```python
categorical_cols = train_df.select_dtypes(
    include=["object"]
).columns

for col in categorical_cols:

    train_df[col].fillna(
        "None",
        inplace=True
    )

    test_df[col].fillna(
        "None",
        inplace=True
    )
```

**📌 PART 08 — Feature Engineering & Data Preprocessing**

**📂 Step 1 — Create New Features**

*Let's first create some meaningful features. A newly built house generally costs more than an old house.*

```python
# House Age
train_df["HouseAge"] = train_df["YrSold"] - train_df["YearBuilt"]
test_df["HouseAge"] = test_df["YrSold"] - test_df["YearBuilt"]
```

**📂 Step 2 — Remodel Age**

*Sometimes a house is renovated. That matters.*

```python
# Remodel Age
train_df["RemodelAge"] = train_df["YrSold"] - train_df["YearRemodAdd"]
test_df["RemodelAge"] = test_df["YrSold"] - test_df["YearRemodAdd"]
```

**📂 Step 3 — Total Bathrooms**

*The dataset separates bathrooms. The model shouldn't have to add them itself.*

```python
train_df["TotalBathrooms"] = (
    train_df["FullBath"]
    + 0.5 * train_df["HalfBath"]
    + train_df["BsmtFullBath"]
    + 0.5 * train_df["BsmtHalfBath"]
)

test_df["TotalBathrooms"] = (
    test_df["FullBath"]
    + 0.5 * test_df["HalfBath"]
    + test_df["BsmtFullBath"]
    + 0.5 * test_df["BsmtHalfBath"]
)
```

**📂 Step 4 — Total Living Area**

```python
train_df["TotalSF"] = (
    train_df["TotalBsmtSF"]
    + train_df["1stFlrSF"]
    + train_df["2ndFlrSF"]
)

test_df["TotalSF"] = (
    test_df["TotalBsmtSF"]
    + test_df["1stFlrSF"]
    + test_df["2ndFlrSF"]
)
```

**📂 Step 5 — Total Porch Area**

```python
train_df["TotalPorchSF"] = (
    train_df["OpenPorchSF"]
    + train_df["EnclosedPorch"]
    + train_df["3SsnPorch"]
    + train_df["ScreenPorch"]
)

test_df["TotalPorchSF"] = (
    test_df["OpenPorchSF"]
    + test_df["EnclosedPorch"]
    + test_df["3SsnPorch"]
    + test_df["ScreenPorch"]
)
```

**📂 Step 6 — Check Newly Created Features**

```python
train_df[[
    "HouseAge",
    "RemodelAge",
    "TotalBathrooms",
    "TotalSF",
    "TotalPorchSF"
]].head()
```

**📂 Step 7 — Target Variable Log Transformation**

```python
import numpy as np

train_df["SalePrice"] = np.log1p(train_df["SalePrice"])
```

**📂 Step 8 — One-Hot Encoding**

```python
train_df = pd.get_dummies(train_df)

test_df = pd.get_dummies(test_df)
```

**📂 Step 9 — Align Train & Test**

```python
train_df, test_df = train_df.align(
    test_df,
    join="left",
    axis=1,
    fill_value=0
)
```

**📂 Step 10 — Verify Shapes**

```python
print("Train Shape :", train_df.shape)

print("Test Shape  :", test_df.shape)
```

**📌 PART 09 — Model Building**

**Step 1 — Separate Features and Target**

```python
# Features
X = train_df.drop("SalePrice", axis=1)

# Target
y = train_df["SalePrice"]
```

**Step 2 — Train Validation Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

**Step 3 — Check Shapes**

```python
print("X_train :", X_train.shape)
print("X_valid :", X_valid.shape)

print("y_train :", y_train.shape)
print("y_valid :", y_valid.shape)
```

**Step 4 — Import Linear Regression**

```python
from sklearn.linear_model import LinearRegression
```

**Step 5 — Create Model**

```python
baseline_model = LinearRegression()
```

**Step 6 — Train the Model**

```python
baseline_model.fit(
    X_train,
    y_train
)
```

**Step 7 — Make Predictions**

```python
# validation prediction
y_pred = baseline_model.predict(X_valid)
```

```python
test_df_prediction = test_df.drop('SalePrice', axis=1)
# test prediction
test_predictions = baseline_model.predict(test_df_prediction)
```

**Step 8 — Reverse Log Transformation**

```python
import numpy as np

test_predictions = np.expm1(test_predictions)

validation_predictions = np.expm1(y_pred)

actual_validation = np.expm1(y_valid)
```

**Step 9 — Quick Look at Predictions**

```python
prediction_df = pd.DataFrame({

    "Actual": actual_validation,

    "Predicted": validation_predictions

})

prediction_df.head()
```

**📌 PART 10 — Model Evaluation**

**Step 1 — Import Evaluation Metrics

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

import numpy as np
```

**Step 2 — Mean Absolute Error (MAE)**

```python
mae = mean_absolute_error(
    actual_validation,
    validation_predictions
)

print("MAE :", mae)
```

**Step 3 — Mean Squared Error (MSE)**

```python
mse = mean_squared_error(
    actual_validation,
    validation_predictions
)

print("MSE :", mse)
```

**Step 4 — Root Mean Squared Error (RMSE)**

```python
rmse = np.sqrt(mse)

print("RMSE :", rmse)
```

**Step 5 — R² Score**

```python
r2 = r2_score(
    actual_validation,
    validation_predictions
)

print("R² Score :", r2)
```

**Step 6 — Print Everything Together**

```python
print("="*40)

print("Baseline Model Performance")

print("="*40)

print(f"MAE  : {mae:.2f}")

print(f"MSE  : {mse:.2f}")

print(f"RMSE : {rmse:.2f}")

print(f"R²   : {r2:.4f}")
```

**Step 7 — Cross Validation**

```python
# Instead of trusting only one train-test split, professionals use Cross Validation.
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(

    baseline_model,

    X,

    y,

    cv=5,

    scoring="neg_root_mean_squared_error"

)
```

```python
cv_rmse = -cv_scores

print(cv_rmse)
```

```python
print("Average CV RMSE :", cv_rmse.mean())

print("Standard Deviation :", cv_rmse.std())
```

**📌 PART 11 — Advanced Regression Models & Model Comparison**

**Step 1 — Import Models**

```python
from sklearn.linear_model import (
    LinearRegression,
    Ridge,
    Lasso,
    ElasticNet
)

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import (
    RandomForestRegressor,
    GradientBoostingRegressor
)
```

**Step 2 — Create Models**

```python
models = {

    "Linear Regression": LinearRegression(),

    "Ridge Regression": Ridge(alpha=1.0),

    "Lasso Regression": Lasso(alpha=0.0005),

    "ElasticNet": ElasticNet(
        alpha=0.0005,
        l1_ratio=0.9
    ),

    "Decision Tree": DecisionTreeRegressor(
        random_state=42
    ),

    "Random Forest": RandomForestRegressor(
        n_estimators=200,
        random_state=42
    ),

    "Gradient Boosting": GradientBoostingRegressor(
        random_state=42
    )

}
```

**Step 3 — Create Empty Results List**

```python
results = []
```

**Step 4 — Train Every Model**

```python
for name, model in models.items():

    model.fit(
        X_train,
        y_train
    )

    predictions = model.predict(X_valid)

    predictions = np.expm1(predictions)

    actual = np.expm1(y_valid)

    mae = mean_absolute_error(
        actual,
        predictions
    )

    mse = mean_squared_error(
        actual,
        predictions
    )

    rmse = np.sqrt(mse)

    r2 = r2_score(
        actual,
        predictions
    )

    results.append({

        "Model": name,

        "MAE": mae,

        "RMSE": rmse,

        "R²": r2

    })
```

**Step 5 — Create Comparison Table**

```python
results_df = pd.DataFrame(results)
```

```python
results_df = results_df.sort_values(
    by="RMSE"
)
```

```python
results_df
```

**Step 6 — Best Model**

```python
best_model_name = results_df.iloc[0]["Model"]

print("Best Model :", best_model_name)
```

**Step 7 — Train Best Model on Full Training Data**

```python
best_model = models[best_model_name]

best_model.fit(
    X_train,
    y_train
)
```

**Step 8 — Feature Importance (Tree Models Only)**

```python
# If the best model is Random Forest, Decision Tree, or Gradient Boosting:
if hasattr(best_model, "feature_importances_"):

    feature_importance = pd.DataFrame({

        "Feature": X_train.columns,

        "Importance": best_model.feature_importances_

    })

    feature_importance = feature_importance.sort_values(
        by="Importance",
        ascending=False
    )

    print(feature_importance.head(15))
```
